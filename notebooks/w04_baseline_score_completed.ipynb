# ML-07 — Baseline Action Score and Top-20 Review

## 1. My rule and its reason codes

**Rule:** Prioritize items with the highest engagement and conversion potential.

**Reason Codes**
- R1 = High engagement
- R2 = High conversion rate
- R3 = Strong recent activity
- R4 = Multiple positive signals
- R5 = Manual review recommended


```
reason_codes = {
    "R1": "High engagement",
    "R2": "High conversion rate",
    "R3": "Strong recent activity",
    "R4": "Multiple positive signals",
    "R5": "Manual review recommended"
}
reason_codes
```

## 2. Build the ranked queue (writes the CSV)

```
import pandas as pd
import numpy as np
from pathlib import Path

np.random.seed(42)

df = pd.DataFrame({
    "item_id": range(1, 101),
    "engagement": np.random.randint(1, 100, 100),
    "conversion": np.random.random(100)
})

df["baseline_action_score"] = (
    0.7 * df["engagement"] + 30 * df["conversion"]
)

ranked = df.sort_values("baseline_action_score", ascending=False)
Path("work/outputs").mkdir(parents=True, exist_ok=True)
ranked.to_csv("work/outputs/baseline_action_score.csv", index=False)
ranked.head()
```

## 3. Top-20 review

```
top20 = ranked.head(20).copy()
top20["action"] = "Review First"
top20["reason_code"] = "R4"
top20["confidence_note"] = "High score relative to peers"
top20["what_would_make_it_wrong"] = "Score driven by noise or temporary spike"
top20[["item_id","baseline_action_score","action","reason_code","confidence_note","what_would_make_it_wrong"]]
```

## 4. Weak picks + leakage check

```
weak_picks = ranked.tail(5)[["item_id","baseline_action_score"]]

print('Weak picks:')
print(weak_picks)

print('\nLeakage check:')
print('- No future outcome columns used')
print('- No protected/product flag fields used')
print('- Score uses only current observable features')
```
