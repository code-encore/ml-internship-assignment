## 1. My lane as an ML task (type)

**Task Type: Ranking**

My lane is content refresh prioritization. The goal is not simply to classify pages as good or bad. Instead, the goal is to rank eligible pages by their likelihood of benefiting from a content refresh.

A ranking approach is appropriate because multiple pages may qualify for refresh opportunities, but resources are limited. The model should help prioritize which pages deserve attention first based on observed search-performance signals.


```python
import pandas as pd

df = pd.read_csv("content_refresh_anonymized(2).csv")
print("Rows:", len(df))
print("Columns:", len(df.columns))
```

## 2. Target or proxy

**Target / Proxy Label:** Refresh Opportunity Score

The dataset does not contain a direct outcome showing whether a refresh was successful. Therefore, a proxy label is required.

A practical proxy is based on observed decline signals such as:

- Downward traffic trend
- High impressions with declining performance
- Page-one ranking decay risk
- Existing search demand

Pages exhibiting these signals can be assigned higher refresh-priority scores. This creates a decision-support target derived from observed behavior rather than a guaranteed business outcome.


```python
eligible = df[(df["impressions_90d"] > 0) & (df["content_age_days"] >= 90)]

print("Eligible pages:", len(eligible))
```

## 3. Success metric

**Primary Metric: Precision@K**

The business goal is to identify the highest-value refresh opportunities.

Precision@K measures how many of the top-ranked recommendations actually satisfy the chosen refresh-opportunity criteria.

A useful target would be maintaining strong precision among the top recommendations reviewed by content teams. This aligns model performance with real operational decisions.


```python
total = len(eligible)
down_pct = (eligible["trend_direction"].eq("down").mean()) * 100

print(f"Downward trend pages: {down_pct:.1f}%")
```

## 4. The unit of analysis, as a real dataframe

**One row = one content page**

Each row represents a single piece of content with associated search-performance signals, rankings, impressions, clicks, and trend indicators.

The model evaluates pages independently and assigns a refresh-priority score or rank.


```python
display(df.head())

print("\nUnit of analysis: One row = one content page")
```

## 5. Why ML beats a fixed rule here

A fixed rule may capture one signal but ignores interactions between many signals.

For example:

- A page may have declining clicks but strong impressions.
- Another page may rank on page one but show early signs of decay.
- Some pages have overlapping opportunity signals.

These patterns are multidimensional and vary across clients and content types. Ranking models can combine many signals simultaneously and learn which combinations indicate the strongest refresh opportunities. This makes the prioritization process more flexible and scalable than a single if-statement.
