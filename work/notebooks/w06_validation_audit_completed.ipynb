## 1. Two paper findings + my methodology questions

### Finding 1
The paper reports improved performance compared with a baseline approach.

**Methodology question:** Where does the label come from?

If labels are derived from information that overlaps with model inputs, measured performance can be inflated. Understanding label creation helps determine whether the model is learning useful patterns or recovering information already present in the features.

### Finding 2
The paper reports that the model generalizes across different clients and content types.

**Methodology question:** Does the validation design support the claim?

If examples from the same client appear in both training and testing data, performance may overestimate real-world generalization. Grouped validation provides stronger evidence of transfer to unseen clients.
