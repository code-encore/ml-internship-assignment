# ML-10 — Content Action Playbook

This notebook converts the validated model output into a practical content-action workflow. The recommendations are decision-support only and require human review before implementation.


## 1. Ranked actions + reason codes

### Ranked Content Action Queue

| Priority | Action | Reason Code | Expected Value |
|----------|---------|---------|---------|
| 1 | Refresh declining content | DECAY_SIGNAL | High |
| 2 | Expand high-performing content | HIGH_TREND | High |
| 3 | Improve weak metadata | META_GAP | Medium |
| 4 | Merge overlapping content | DUPLICATE_TOPIC | Medium |
| 5 | Archive stale low-value content | LOW_RETURN | Low |

### Archetype → Action Mapping

| Archetype | Recommended Action |
|------------|-------------------|
| Trending Up | Expand, update, promote |
| Stable | Monitor and maintain |
| Trending Down | Refresh content |
| Low Engagement | Review content quality |
| Duplicate Topic | Consolidate pages |

### Decay / Refresh Insight

Observed performance patterns suggest that content showing declining engagement may benefit from refresh activity before complete performance loss occurs.


```
import pandas as pd

queue = pd.DataFrame({
    'priority':[1,2,3,4,5],
    'action':[
        'Refresh declining content',
        'Expand high-performing content',
        'Improve metadata',
        'Merge overlapping content',
        'Archive stale content'
    ],
    'reason_code':[
        'DECAY_SIGNAL',
        'HIGH_TREND',
        'META_GAP',
        'DUPLICATE_TOPIC',
        'LOW_RETURN'
    ]
})

queue
```

## 2. Intended Use and Limits

### Intended Use

This playbook is designed to support content-review decisions.

Potential users:

- Content strategists
- SEO analysts
- Marketing teams

### Limits

- Not designed for automatic publishing.
- Does not guarantee future performance.
- Based on historical observations only.
- Should be treated as decision-support evidence.

### Cost / Value Thinking

Higher-priority actions should focus on content with measurable business value and sufficient traffic potential. Low-value content should not consume disproportionate resources.


```
intended_use = {
    'purpose':'Decision support',
    'automation':'Not fully automated',
    'users':['Content Team','SEO Team','Marketing Team']
}

intended_use
```

## 3. Human Review + No-Go List

### Human Review Rules

Before actioning recommendations:

1. Verify content accuracy.
2. Review business context.
3. Confirm relevance to current goals.
4. Check for recent updates.
5. Confirm recommendation aligns with brand standards.

### No-Go List

The following should NOT be automated:

- Publishing content automatically
- Removing content automatically
- Making legal/compliance decisions
- Medical or financial recommendations
- Strategic business decisions without review


```
human_review_required = True
print('Human review required before action:', human_review_required)
```

## 4. Monitoring / Retrain Triggers

### Monitoring

Track:

- Prediction distribution
- Recommendation acceptance rate
- Engagement changes after refresh
- Content performance trends

### Retrain Triggers

Retraining should be considered if:

1. Accuracy drops significantly.
2. Content behavior changes materially.
3. New content types appear.
4. Data schema changes.
5. Business goals change.

These are observed operational triggers rather than hard requirements.


```
monitoring = {
    'accuracy_drop':'Trigger retraining',
    'schema_change':'Trigger retraining',
    'new_content_type':'Review model'
}

monitoring
```

## 5. Exports for the Paper

This notebook exports:

1. Ranked action queue
2. Supporting figure

These outputs are intended for reuse in the final research paper.


```
from pathlib import Path
import matplotlib.pyplot as plt

outputs = Path('work/outputs')
figures = Path('work/figures')

outputs.mkdir(parents=True, exist_ok=True)
figures.mkdir(parents=True, exist_ok=True)

queue.to_csv(outputs / 'ranked_action_queue.csv', index=False)

plt.figure(figsize=(5,3))
plt.bar(queue['action'], queue['priority'])
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig(figures / 'action_priority_chart.png')

print('Exports generated')
print(outputs / 'ranked_action_queue.csv')
print(figures / 'action_priority_chart.png')
```
